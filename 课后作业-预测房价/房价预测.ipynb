```python
import numpy as np
df=pd.concat([data, pd.get_dummies(data['CHAS'])], axis = 1 ) #CHAS靠近河床类别 建立虚拟变量
del df[1.0]  #删除冗余的变量
del df['CHAS']  #删除冗余的变量
#df[0.0].value_counts()
df[0.0]=df[0.0].astype(np.float64)
df.columns=['CRIM', 'ZN','INDUS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','PRICE','0.0']
```

```python
'''使用多种方法预测'''
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

y= df['PRICE'].values
X = df[['CRIM','ZN','INDUS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','0.0']]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state = 33)

print('多元线性回归')
regr = LinearRegression()#'''多元线性回归'''
regr.fit(X_train, y_train)
print(regr.score(X_train, y_train))
regr.fit(X_test, y_test)
print(regr.score(X_test, y_test))
print('knn')
knn = KNeighborsRegressor() #knn
knn.fit(X_train,y_train)
print(knn.score(X_train, y_train))
knn.fit(X_test, y_test)
print(knn.score(X_test, y_test))
print('岭回归')
ridge = Ridge()#岭回归
ridge.fit(X_train,y_train)
print(ridge.score(X_train, y_train))
ridge.fit(X_test, y_test)
print(ridge.score(X_test, y_test))
print('lasso回归')
lasso = Lasso()#lasso回归
lasso.fit(X_train,y_train)
print(lasso.score(X_train, y_train))
lasso.fit(X_test, y_test)
print(lasso.score(X_test, y_test))
print('回归决策树')
decision = DecisionTreeRegressor()#回归决策树，结果过拟合
decision.fit(X_train,y_train)
print(decision.score(X_train, y_train))
decision.fit(X_test, y_test)
print(decision.score(X_test, y_test))
print('支持向量机回归')
svr = SVR()  #支持向量机回归
svr.fit(X_train,y_train)
print(svr.score(X_train, y_train))
svr.fit(X_test, y_test)
print(svr.score(X_test, y_test))
```

```python
import statsmodels.api as sm#最小二乘法
X2 = sm.add_constant(X)#增加常数项
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())  
#R-squared拟合度优度0.741，F检验P值小于0.05，对房价有显著性影响
```
